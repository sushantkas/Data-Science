```python
x=train_tranform.drop("Transported", axis=1)
y=train_tranform["Transported"]
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
x_train, x_test, y_train, y_test=train_test_split(x,y ,test_size=0.10, random_state=42)
logic_lodel=LogisticRegression(solver='lbfgs', max_iter=1000)
forest_model=RandomForestClassifier(random_state=0, n_estimators=931, max_depth=5)
```

```python
logic_lodel.fit(x_train, y_train)
```

```python
logic_lodel.score(x_test,y_test)
```

```python
forest_model.fit(x_train,y_train)
forest_model.score(x_test,y_test)
```
